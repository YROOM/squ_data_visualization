```python
# 1. 导入三大件
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. 设置美学参数（这是Seaborn的精髓之一）
sns.set_theme(style="whitegrid")  # 设置背景风格：白色网格

# 3. 加载内置数据集
tips = sns.load_dataset("tips")
print(tips.head())  # 快速查看：total_bill, tip, sex, smoker, day, time, size
```

```python
plt.figure(figsize=(12, 4))

# 单独看一列数据，数据长什么样？
# 直方图 + 核密度图 —— histplot / displot
plt.subplot(1, 2, 1)
# 直方图：查看账单金额分布
sns.histplot(data=tips, x='total_bill', bins=50, kde=True)
plt.title('hist +KDE')

plt.subplot(1, 2, 2)
# 单纯的密度图（平滑曲线）
sns.kdeplot(data=tips, x='total_bill', fill=True)
plt.title('value')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
# 查看星期几的客人最多
sns.countplot(data=tips, x='day', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('daily person')
plt.show()
```

```python
# 双变量分析——关系与对比（20分钟）
plt.figure(figsize=(8, 6))
# 分析：账单金额 vs 小费金额
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('total bill and tips')
plt.show()
```

```python
# 快速演示：加上颜色（hue）看性别差异
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python

# 不使用 hue - 所有点都是同一种颜色
sns.relplot(data=tips, x='total_bill', y='tip')
plt.show()
```
