```python
import matplotlib.pyplot as plt
import numpy as np
N = 60
np.random.seed(100) #设置随机数种子
x = np.random.rand(N)
y = np.random.rand(N)
s = np.pi*(15*np.random.rand(N))**2
c =-s
opacity = 0.5
fig, axes = plt.subplots()
axes.scatter(x, y,s,c,alpha = opacity)
plt.show()
```

# 下面这个是个饼图的示例

```python
fig, axes = plt.subplots()
labels = ['Taxi', 'Metro', 'Walk', 'Bus', 'Bicycle', 'Driving']
sizes = [10, 30, 5, 25, 5, 25]
explode = (0, 0.1, 0, 0, 0, 0)
axes.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
axes.axis('equal')
axes.set_title('pie chart')
plt.show()
```
